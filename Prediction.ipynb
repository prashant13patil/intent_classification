```python
from nltk import word_tokenize
from tensorflow.keras.preprocessing.text import Tokenizer
maxlen = 25
new_complaint = [str(input("enter text"))]
#seq = new_complaint.texts_to_sequences(new_complaint)
#new_complaint = [new_complaint]
print(new_complaint)
#tokenizer= Tokenizer()
#tokenizer.fit_on_texts(new_complaint)

seq = train.tokenizer.texts_to_sequences(new_complaint)
seq
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded = pad_sequences(seq, maxlen=25,padding= "pre")

padded
```

```python
# transform to required shape
def transform_x(data, tokenizer):
    output_shape= [data.shape[0],
                  data.shape[1],
                  train.tokenizer.word_index.keys().__len__()]
    results= np.zeros(output_shape)
    
    for i in range(data.shape[0]):
        for ii in range(data.shape[1]):
            results[i, ii, data[i,ii]-1]= 1
    return results

xtr_transformed= transform_x(padded, train.tokenizer)
xtr_transformed.shape
```
